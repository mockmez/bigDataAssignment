```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Read HDFS File") \
    .config("spark.hadoop.fs.defaultFS", "hdfs://localhost:9010") \
    .getOrCreate()

```

```python
books_and_genres_df = spark.read.csv(
    "hdfs://localhost:9010/user/mango/dataset/books_and_genres.csv",
    header=True,
    inferSchema=True,
    multiLine=True,
    escape="\"",
    sep=",",
    quote='"'
)

books_df = spark.read.csv(
    "hdfs://localhost:9010/user/mango/dataset/Books.csv",
    header=True,
    inferSchema=True
)

users_df = spark.read.csv(
    "hdfs://localhost:9010/user/mango/dataset/Users.csv",
    header=True,
    inferSchema=True
)
```
